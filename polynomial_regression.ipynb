```python

# This sets up the graphing configuration
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (15,5)
plt.rcParams["font.family"] = "DejaVu Sans"
plt.rcParams["font.size"] = "12"
plt.rcParams['image.cmap'] = 'rainbow'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['figure.facecolor'] = 'white'
import numpy as np
import pandas as pd

dataset = pd.read_csv('../data/traffic_by_hour.csv')
print(dataset.head())
```

```python
dataset_T = np.transpose(dataset)
print (dataset_T.head())
```

```python
for sample in range (0,dataset_T.shape[1]):
    plt.plot(dataset.columns.values,dataset_T[sample])
plt.xlabel("Time of day")
plt.ylabel("internet traffice(Gps)")
plt.show()
```

```python
hours = dataset.columns.values
train_X = np.transpose([int(hour) for hour in hours ])
train_Y = [dataset[hour].mean() for hour in hours]

plt.scatter(train_X,train_Y)

for sample in range(0,dataset_T.shape[1]):
    plt.plot(hours, dataset_T[sample], alpha=0.25)
plt.xlabel('Time in day')
plt.ylabel('Internet traffice (Gps)')    

for sample in range(0,dataset_T.shape[1]):
    plt.plot(hours,dataset_T[sample],alpha=0.25)
plt.xlabel('time of day')
plt.ylabel('internet traffice (Gps) ')    
plt.show()
```

```python
poly_1 = np.polyfit(train_X,train_Y,1)
poly_2 = np.polyfit(train_X,train_Y,2)
poly_3 = np.polyfit(train_X,train_Y,3)
poly_4 = np.polyfit(train_X,train_Y,4)
plt.scatter(train_X,train_Y)
xp = np.linspace(0,24,100)

plt.plot(xp,np.polyval(poly_1,xp),'k--')
plt.plot(xp,np.polyval(poly_2,xp),'r--')
plt.plot(xp,np.polyval(poly_3,xp),'b--')
plt.plot(xp,np.polyval(poly_4,xp),'y--')

plt.xticks(train_X,dataset.columns.values)
plt.xlabel('Time of day')
plt.ylabel('Internet traffice (Gps)')
plt.show()
```
